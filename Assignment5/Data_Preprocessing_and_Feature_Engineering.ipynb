```python
# Display the columns with null values
null_columns = df.columns[df.isnull().any()]
print("Columns with null values:\n", null_columns)
```

```python
# Fill null values with appropriate strategies
# Numerical columns: fill with mean
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_cols] = df[numerical_cols].apply(lambda x: x.fillna(x.mean()), axis=0)

# Categorical columns: fill with mode
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].apply(lambda x: x.fillna(x.mode()[0]), axis=0)

# For specific columns where null values have special meaning
df['Alley'] = df['Alley'].fillna('No Alley')
df['PoolQC'] = df['PoolQC'].fillna('No Pool')
df['Fence'] = df['Fence'].fillna('No Fence')
df['MiscFeature'] = df['MiscFeature'].fillna('No Misc Feature')
```
